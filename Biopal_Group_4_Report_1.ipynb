###**3.4.1 Justification of biomarkers**###

*Fasting Glucose*

Fasting glucose is the measurement of blood glucose levels after a period of no food or drink (typically 8-12 hours overnight). It provides a direct snapshot of blood glucose levels after an overnight fast, reflecting the body's immediate ability to regulate sugar, primarily driven by hepatic glucose production and basal insulin sensitivity. When the pancreas does not produce enough insulin, or cells do not respond effectively to it, glucose remains in the bloodstream, leading to elevated fasting levels. As such, Fasting glucose is a primary diagnostic criterion for prediabetes and Type 2 Diabetes, with higher values predicting increased future risk.

*HbA1c*

HbA1c measures the average blood glucose level over the preceding two to three months. This long-term measure occurs when glucose in the bloodstream non-enzymatically attaches to hemoglobin within red blood cells, a process proportional to the ambient blood glucose concentration. Persistently high blood sugar leads to greater glycation, resulting in an elevated HbA1c percentage. Consequently, HbA1c serves as a key diagnostic tool for diabetes and is a strong predictor of the risk for developing long-term microvascular and macrovascular complications.

*Triglyceride*

Triglyceride are a type of fat in the blood that serve as a crucial energy source. Their levels are significantly influenced by dietary intake, particularly refined carbohydrates and unhealthy fats, and are intimately linked to insulin resistance. When cells become resistant to insulin, the liver may increase its production of triglycerides and very-low-density lipoproteins (VLDL), leading to elevated circulating levels. As such, high triglycerides are a core component of metabolic syndrome and represent an independent risk factor for cardiovascular disease, which frequently co-occurs with Type 2 Diabetes.

To conclude, the true power of using all three biomarkers in combination lies in their ability to provide a comprehensive, integrated assessment of metabolic health. Fasting glucose offers current control, HbA1c reveals long-term glycemic history, and triglycerides shed light on lipid metabolism and underlying insulin resistance. Together, they allow for more accurate risk stratification, enabling the earlier identification of individuals at high risk for Type 2 Diabetes and its associated complications, even if no single marker overtly indicates a problem. This integrated approach is essential for understanding the full scope of metabolic dysfunction and for tailoring effective lifestyle and clinical interventions.

#### **5.2.1 Define smoking status categories**




```python
def categorize_smoking_status(row):
    smq020 = row['SMQ020']
    smq040 = row['SMQ040']

    # Never Smoker: SMQ020 is 2 (No)
    if smq020 == 2:
        return "Never Smoker"
    # Former Smoker: SMQ020 is 1 (Yes) and SMQ040 is 3 (Now smokes some days) or 2 (Now smokes every day)
    elif smq020 == 1 and (smq040 == 3 or smq040 == 2):
        # Note: The provided rule for Former Smoker seems to overlap with Current Smoker based on SMQ040=2.
        # Assuming the intent is: Former Smoker if SMQ020=1 and SMQ040 is NOT 1 (Every day) or 2 (Some days),
        # and the participant answered Yes to SMQ020 (smoked at least 100 cigarettes in life).
        # A more common definition for Former Smoker is SMQ020=1 and SMQ040=3 (Smokes some days) or SMQ040=2 (Smokes every day) is incorrect.
        # Based on NHANES documentation, SMQ040 codes are usually:
        # 1 = Every day
        # 2 = Some days
        # 3 = Not at all
        # So, the rule for Former Smoker should be SMQ020=1 and SMQ040=3 (Not at all).
        # And Current Smoker should be SMQ020=1 and (SMQ040=1 or SMQ040=2).
        # I will follow the likely intended logic based on standard smoking status definitions and SMQ040 values (1=Every day, 2=Some days, 3=Not at all).
        if smq040 == 3:
             return "Former Smoker"
        elif smq040 == 1 or smq040 == 2:
             return "Current Smoker"
        else:
             return "Unknown" # Handle cases where SMQ020 is 1 but SMQ040 is unexpected
    # Current Smoker: SMQ020 is 1 (Yes) and SMQ040 is 1 (Every day) or 2 (Some days)
    # This condition is covered in the previous elif based on the corrected understanding of SMQ040.

    # Handle other cases like NaN in SMQ020
    return "Unknown"


df_smoking_merged['Smoking_Status'] = df_smoking_merged.apply(categorize_smoking_status, axis=1)

print("\nValue Counts for Smoking Status:")
print(df_smoking_merged['Smoking_Status'].value_counts(dropna=False))
```

```python

# Calculate sample size for valid and invalid data points for Smoking status

# Valid data points for Smoking_Status are any non-'Unknown' values.
# Invalid data points are 'Unknown' or NaN values in the new 'Smoking_Status' column.
# Note: The `categorize_smoking_status` function currently returns 'Unknown' for NaN SMQ020 values and unexpected SMQ040 values when SMQ020 is 1.
# We can consider NaN in the original SMQ columns as contributing to 'invalid' if they lead to 'Unknown' status.
# However, the request specifically asks for valid/invalid data points *for Smoking status*.
# The most direct way is to count the values in the resulting 'Smoking_Status' column.

# Valid data points for the derived 'Smoking_Status' are the sum of 'Never Smoker', 'Former Smoker', and 'Current Smoker'.
valid_smoking_status = df_smoking_merged[df_smoking_merged['Smoking_Status'] != 'Unknown']['Smoking_Status'].count()

# Invalid data points for 'Smoking_Status' are the count of 'Unknown' values and any potential NaNs if not handled by the function.
invalid_smoking_status = df_smoking_merged[df_smoking_merged['Smoking_Status'] == 'Unknown']['Smoking_Status'].count()
# Check for any NaNs in the final Smoking_Status column (should be 0 if 'Unknown' handles NaNs in SMQ020)
invalid_smoking_status += df_smoking_merged['Smoking_Status'].isna().sum()


print("\nSample Size for Smoking Status:")
print(f"Valid: {valid_smoking_status}")
print(f"Invalid: {invalid_smoking_status}")
print(f"Total: {valid_smoking_status + invalid_smoking_status}") # Should match the merged DataFrame size
```

```python
import pandas as pd
import numpy as np
import requests # Import requests for robust URL checking
from io import BytesIO # Import BytesIO to read content directly from requests
from tqdm.notebook import tqdm # For progress bars in notebooks/interactive environments

# --- Configuration ---
# Define years and cycles for consistent URL construction
# Note: For 2017-2018 (J) and 2018-2020 (J), the 'year' in the URL is '2017'.
# The actual data cycle code for 2018-2020 is also 'J'.
# NHANES data is released in 2-year cycles. The "2020" you have is
# effectively the 2017-2018 data due to COVID-19 related changes and
# continuous release. For consistency with how CDC names files,
# we'll map years to the directory year and cycle code correctly.

# NHANES URL mapping: {display_year_range: {'url_year': 'year_in_url', 'file_prefix': 'cycle_code'}}
# I'm using the standard 2-year cycle notation which is clearer for NHANES.
NHANES_CYCLES = {
    "2007-2008": {'url_year': '2007', 'file_prefix': 'E'},
    "2009-2010": {'url_year': '2009', 'file_prefix': 'F'},
    "2011-2012": {'url_year': '2011', 'file_prefix': 'G'},
    "2013-2014": {'url_year': '2013', 'file_prefix': 'H'},
    "2015-2016": {'url_year': '2015', 'file_prefix': 'I'},
    "2017-2018": {'url_year': '2017', 'file_prefix': 'J'},
    # The 2018-2020 (post-COVID) data often references 2017 files for consistency
    # as no new full cycle data was released. It uses the 'J' prefix.
    # If your analysis specifically needs to distinguish, you'd need to consult
    # specific NHANES documentation for this specific data release.
    # For strict 2017-2020 data, NHANES usually calls it "Continuous NHANES 2017-March 2020 Pre-Pandemic File"
    # and it uses the 'J' cycle code for files.
    "2017-2020": {'url_year': '2017', 'file_prefix': 'J'} # Using '2017-2020' as a key for this range.
}


# --- Helper Function for Robust URL Reading ---
def read_nhanes_xpt(url, cols_to_select=None):
    """
    Reads an XPT file directly from a URL using requests and pandas.read_sas.
    Handles potential file not found errors and selects specified columns.
    Explicitly sets format='xport'.
    """
    try:
        response = requests.get(url)
        response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
        # Use BytesIO to read content directly from the response
        df = pd.read_sas(BytesIO(response.content), format='xport') # Explicitly set format
        if cols_to_select:
            # Select only columns that exist in the DataFrame
            existing_cols = [col for col in cols_to_select if col in df.columns]
            if len(existing_cols) < len(cols_to_select):
                print(f"Warning: Some requested columns not found in {url.split('/')[-1]}: {set(cols_to_select) - set(existing_cols)}")
            df = df[existing_cols]
        return df
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {url}: {e}")
        return None
    except Exception as e:
        print(f"Error reading XPT from {url}: {e}")
        return None

# --- Data Processing Functions ---

# Removed process_liver_data function
# Removed process_demographics function

# Removed process_bmi_data function

def process_sleep_data(df_sleep_raw):
    if df_sleep_raw is None or df_sleep_raw.empty:
        return pd.DataFrame()

    df_sleep = df_sleep_raw.copy()
    # Prioritize SLD012/SLD013 as discussed for average sleep duration
    # SLD010H was present in older cycles, but SLD012/SLD013 are more common in recent ones.
    # For this modified script, we'll favor the SLD012/SLD013 average approach.
    sleep_cols_to_check = ['SEQN', 'SLD012', 'SLD013', 'SLD010H'] # Order of preference, include SEQN

    # Convert to numeric, handle NHANES missing codes
    for col in sleep_cols_to_check:
        if col in df_sleep.columns and col != 'SEQN': # Only convert non-SEQN columns
            df_sleep[col] = pd.to_numeric(df_sleep[col], errors='coerce').replace({7: np.nan, 9: np.nan, 77: np.nan, 99: np.nan})

    df_sleep['Avg_Sleep_Hours'] = np.nan

    if 'SLD012' in df_sleep.columns and 'SLD013' in df_sleep.columns:
        # Calculate average of weekday and weekend sleep
        df_sleep['Avg_Sleep_Hours'] = df_sleep[['SLD012', 'SLD013']].mean(axis=1)
    elif 'SLD012' in df_sleep.columns: # Fallback if only weekday sleep is available/valid
        df_sleep['Avg_Sleep_Hours'] = df_sleep['SLD012']
    elif 'SLD010H' in df_sleep.columns: # Older cycle alternative
        df_sleep['Avg_Sleep_Hours'] = df_sleep['SLD010H']
    else:
        print(f"Warning: No valid sleep duration variable found in this cycle's SLQ data.")

    return df_sleep[['SEQN', 'Avg_Sleep_Hours']].copy()

# Removed process_alcohol_data function


def process_physical_activity_data(df_pa_raw):
    """
    Processes raw PAQ data to calculate weekly vigorous recreational physical activity.
    """
    if df_pa_raw is None or df_pa_raw.empty:
        return pd.DataFrame()

    df_pa = df_pa_raw.copy()
    pa_cols = ['SEQN', 'PAQ650', 'PAQ655', 'PAD660'] # Screener, Days, Minutes

    for col in pa_cols:
        if col in df_pa.columns and col != 'SEQN': # Only convert non-SEQN columns
            df_pa[col] = pd.to_numeric(df_pa[col], errors='coerce')
            # Replace common NHANES missing codes
            if col == 'PAQ650': df_pa[col] = df_pa[col].replace({7: np.nan, 9: np.nan})
            elif col == 'PAQ655': df_pa[col] = df_pa[col].replace({77: np.nan, 99: np.nan})
            elif col == 'PAD660': df_pa[col] = df_pa[col].replace({7777: np.nan, 9999: np.nan})
        elif col != 'SEQN': # If a critical column is missing (and not SEQN), fill with NaN and print warning
            df_pa[col] = np.nan
            print(f"Warning: Missing PAQ column '{col}' for PA processing. May affect results.")


    # Initialize the new calculated column
    df_pa['Vigorous_Rec_Min_Week'] = np.nan # Start with NaN to distinguish from 0

    # Condition for valid vigorous activity data (PAQ650=Yes, and days/minutes are not NaN)
    # Ensure that PAQ655 and PAD660 columns exist before using them
    if 'PAQ650' in df_pa.columns and 'PAQ655' in df_pa.columns and 'PAD660' in df_pa.columns:
        valid_vig_pa_cond = (df_pa['PAQ650'] == 1) & \
                            (~df_pa['PAQ655'].isna()) & \
                            (~df_pa['PAD660'].isna())

        # Perform the calculation for valid rows
        df_pa.loc[valid_vig_pa_cond, 'Vigorous_Rec_Min_Week'] = \
            df_pa['PAQ655'] * df_pa['PAD660']

    # Set to 0.0 for those who explicitly said 'No' to vigorous activity (PAQ650 == 2)
    if 'PAQ650' in df_pa.columns:
         df_pa.loc[df_pa['PAQ650'] == 2, 'Vigorous_Rec_Min_Week'] = 0.0


    return df_pa[['SEQN', 'Vigorous_Rec_Min_Week']].copy()

def process_diet_data(df_dr1tot_raw, df_dr2tot_raw):
    """
    Processes raw DR1TOT and DR2TOT data to calculate average total sugar intake.
    """
    if (df_dr1tot_raw is None or df_dr1tot_raw.empty) and \
       (df_dr2tot_raw is None or df_dr2tot_raw.empty):
        print("DR1TOT and DR2TOT DataFrames are empty or not loaded, skipping diet processing.")
        return pd.DataFrame()

    # Create copies to avoid SettingWithCopyWarning
    dr1 = df_dr1tot_raw[['SEQN', 'DR1TSUGR', 'DR1DRSTZ']].copy() if df_dr1tot_raw is not None and not df_dr1tot_raw.empty else pd.DataFrame()
    dr2 = df_dr2tot_raw[['SEQN', 'DR2TSUGR', 'DR2DRSTZ']].copy() if df_dr2tot_raw is not None and not df_dr2tot_raw.empty else pd.DataFrame()


    # Rename columns for merging
    if not dr1.empty: dr1.rename(columns={'DR1TSUGR': 'Total_Sugar_Day1', 'DR1DRSTZ': 'Recall_Status_Day1'}, inplace=True)
    if not dr2.empty: dr2.rename(columns={'DR2TSUGR': 'Total_Sugar_Day2', 'DR2DRSTZ': 'Recall_Status_Day2'}, inplace=True)


    # Merge the two dataframes. Use 'outer' to keep all SEQN from either day.
    if not dr1.empty and not dr2.empty:
        diet_df = pd.merge(dr1, dr2, on='SEQN', how='outer')
    elif not dr1.empty:
        diet_df = dr1
    elif not dr2.empty:
        diet_df = dr2
    else:
        return pd.DataFrame({'SEQN': [], 'Avg_Total_Sugar_g': []}) # Return empty if neither loaded

    # Convert sugar values and recall status to numeric, handle missing codes
    for col in ['Total_Sugar_Day1', 'Total_Sugar_Day2', 'Recall_Status_Day1', 'Recall_Status_Day2']:
        if col in diet_df.columns:
            # Ensure conversion is only attempted if the column exists
            diet_df[col] = pd.to_numeric(diet_df[col], errors='coerce')
            # Assuming 1.0 means reliable. Non-1.0 status or missing should not be used for average.
            if 'Recall_Status' in col:
                diet_df.loc[diet_df[col].notna() & (diet_df[col] != 1), col] = np.nan # Only set to NaN if not 1 and not already NaN


    # Calculate average total sugar intake
    diet_df['Avg_Total_Sugar_g'] = np.nan

    # Case 1: Both days are reliable
    # Ensure columns exist before using them in conditions
    if 'Total_Sugar_Day1' in diet_df.columns and 'Total_Sugar_Day2' in diet_df.columns and 'Recall_Status_Day1' in diet_df.columns and 'Recall_Status_Day2' in diet_df.columns:
        cond_both_reliable = (~diet_df['Total_Sugar_Day1'].isna()) & (diet_df['Recall_Status_Day1'] == 1.0) & \
                             (~diet_df['Total_Sugar_Day2'].isna()) & (diet_df['Recall_Status_Day2'] == 1.0)
        diet_df.loc[cond_both_reliable, 'Avg_Total_Sugar_g'] = (diet_df['Total_Sugar_Day1'] + diet_df['Total_Sugar_Day2']) / 2


    # Case 2: Only Day 1 is reliable
    # Ensure columns exist before using them in conditions
    if 'Total_Sugar_Day1' in diet_df.columns and 'Recall_Status_Day1' in diet_df.columns and 'Total_Sugar_Day2' in diet_df.columns and 'Recall_Status_Day2' in diet_df.columns:
        cond_day1_only = (~diet_df['Total_Sugar_Day1'].isna()) & (diet_df['Recall_Status_Day1'] == 1.0) & \
                         (diet_df['Total_Sugar_Day2'].isna() | (diet_df['Recall_Status_Day2'].notna() & (diet_df['Recall_Status_Day2'] != 1.0))) # Check if Day2 is NaN or not reliable
        diet_df.loc[cond_day1_only, 'Avg_Total_Sugar_g'] = diet_df['Total_Sugar_Day1']


    # Case 3: Only Day 2 is reliable
    # Ensure columns exist before using them in conditions
    if 'Total_Sugar_Day1' in diet_df.columns and 'Recall_Status_Day1' in diet_df.columns and 'Total_Sugar_Day2' in diet_df.columns and 'Recall_Status_Day2' in diet_df.columns:
        cond_day2_only = (diet_df['Total_Sugar_Day1'].isna() | (diet_df['Recall_Status_Day1'].notna() & (diet_df['Recall_Status_Day1'] != 1.0))) & \
                         (~diet_df['Total_Sugar_Day2'].isna()) & (diet_df['Recall_Status_Day2'] == 1.0) # Check if Day1 is NaN or not reliable
        diet_df.loc[cond_day2_only, 'Avg_Total_Sugar_g'] = diet_df['Total_Sugar_Day2']


    return diet_df[['SEQN', 'Avg_Total_Sugar_g']].copy()


# Removed process_smoking_data function
# Removed process_family_history_data function


# --- Main Data Loading and Merging Process ---

# Use dictionaries to collect dataframes by type from all cycles
processed_data_by_type = {
    'sleep': [],
    'physical_activity': [],
    'diet': []
    # Add other data types here as needed
}


# Map the cycle keys from NHANES_CYCLES for iteration
# This ensures we loop through the defined 2-year cycles including the 2017-2020 special case.
for display_year_range, cycle_info in tqdm(NHANES_CYCLES.items(), desc="Processing NHANES Cycles"):
    url_year = cycle_info['url_year']
    file_prefix = cycle_info['file_prefix']

    print(f"\n--- Loading data for {display_year_range} (URL Year: {url_year}, File Prefix: {file_prefix}) ---")

    try:
        # Define URLs for required files for the current cycle
        # Corrected URLs to use wwwn.cdc.gov and dynamic year/prefix
        file_urls = {
            # Removed DEMO, BMX, BIOPRO and ALQ
            'SLQ':    f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/SLQ_{file_prefix}.XPT",
            'PAQ':    f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/PAQ_{file_prefix}.XPT",
            'DR1TOT': f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/DR1TOT_{file_prefix}.XPT",
            'DR2TOT': f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/DR2TOT_{file_prefix}.XPT",
            # Removed SMQ and DIQ
        }

        # Load raw data for the current cycle
        # Pass the specific columns expected for each file, for efficiency and error handling.
        # Use the read_nhanes_xpt helper function
        # Removed df_demo_raw, df_bmi_raw, df_liver_raw and df_alcohol_raw loading
        df_sleep_raw = read_nhanes_xpt(file_urls['SLQ'], ['SEQN', 'SLD010H', 'SLD012', 'SLD013']) # Include all possible sleep cols
        df_pa_raw = read_nhanes_xpt(file_urls['PAQ'], ['SEQN', 'PAQ650', 'PAQ655', 'PAD660'])
        df_dr1tot_raw = read_nhanes_xpt(file_urls['DR1TOT'], ['SEQN', 'DR1TSUGR', 'DR1DRSTZ'])
        df_dr2tot_raw = read_nhanes_xpt(file_urls['DR2TOT'], ['SEQN', 'DR2TSUGR', 'DR2DRSTZ'])
        # Removed df_smq_raw and df_diq_raw loading


        # Process data using dedicated functions
        # Removed df_demo, df_bmi, df_liver and df_alcohol processing
        df_sleep = process_sleep_data(df_sleep_raw)
        df_pa = process_physical_activity_data(df_pa_raw)
        df_diet = process_diet_data(df_dr1tot_raw, df_dr2tot_raw)
        # Removed df_smoking and df_family_history processing

        # Append processed dataframes to the lists keyed by data type
        if df_sleep is not None and not df_sleep.empty: processed_data_by_type['sleep'].append(df_sleep)
        if df_pa is not None and not df_pa.empty: processed_data_by_type['physical_activity'].append(df_pa)
        if df_diet is not None and not df_diet.empty: processed_data_by_type['diet'].append(df_diet)


        print(f" Loaded and processed data for cycle: {display_year_range}")


    except Exception as e:
        print(f" Failed to process cycle {display_year_range} - Error: {e}")


# --- Merge all processed dataframes and then with df_demo_filtered ---

# Ensure df_demo_filtered from the start of the notebook is available
if 'df_demo_filtered' in locals() and not df_demo_filtered.empty:
    # Start with the user's pre-filtered demographic data
    df_final = df_demo_filtered.copy()
    print("\n--- Merging with User's Filtered Demographic Data ---")
    print("Starting with df_demo_filtered shape:", df_final.shape)

    # Concatenate dataframes for each data type across all cycles
    for data_type, df_list in processed_data_by_type.items():
        if df_list: # Only proceed if there are dataframes to concatenate
            print(f"Concatenating {data_type} data across cycles...")
            df_combined = pd.concat(df_list, ignore_index=True)
            if 'SEQN' in df_combined.columns:
                 print(f"Merging {data_type} data (shape: {df_combined.shape})...")
                 df_final = pd.merge(df_final, df_combined, on="SEQN", how="left")
                 print("Current merged shape:", df_final.shape)
            else:
                 print(f"Warning: Skipping merge for concatenated {data_type} data as it is missing 'SEQN' column.")
        else:
            print(f"No {data_type} data loaded across cycles. Skipping merge.")


    print("\n--- Final Merged Dataset (with user's filtered demographics) ---")
    print(" Final Merged Dataset Shape:", df_final.shape)
    # Removed Cycle distribution as we started with filtered demographics
    print("\nHead of Final Dataset:")
    print(df_final.head())

    # Check and print missing values for key new columns, only if they exist
    key_cols_to_check = ['Vigorous_Rec_Min_Week', 'Avg_Total_Sugar_g', 'Avg_Sleep_Hours'] # Removed smoking and diabetes related columns
    existing_key_cols = [col for col in key_cols_to_check if col in df_final.columns]
    if existing_key_cols:
        print("\nMissing values in key new columns:")
        print(df_final[existing_key_cols].isnull().sum())
    else:
        print("\nNo key processed columns (Physical Activity, Sleep, Diet) found in the final dataset.")

else:
    print(" User's filtered demographic data ('df_demo_filtered') not found or is empty. Please run the initial demographic processing cells.")
```

### **5.7 Descriptive Statistics and Normality Checks for Biomarkers**



```python
from scipy.stats import shapiro

biomarker_dataframes = {
    'Fasting Glucose (mg/dL)': df_glu['LBXGLU'],
    'HbA1c (%)': df_ghb['LBXGH'],
    'Triglyceride (mg/dL)': df_trigly['LBXTR']
}

print("--- Descriptive Statistics for Biomarkers ---")
for name, series in biomarker_dataframes.items():
    print(f"\n{name}:")
    display(series.describe())
    print(f"Skewness: {series.skew():.4f}")


print("\n--- Normality Checks (Shapiro-Wilk Test) ---")
print("Null Hypothesis (H0): The data is drawn from a normal distribution.")
print("Alternative Hypothesis (H1): The data is NOT drawn from a normal distribution.")
print("Interpretation: A p-value < 0.05 leads to rejection of H0, suggesting non-normality.")

for name, series in biomarker_dataframes.items():
    # Shapiro-Wilk is sensitive to sample size, and for large samples,
    # it might reject normality even for small deviations.
    # It also requires non-missing data. We'll drop NaNs for the test.
    series_clean = series.dropna()
    if len(series_clean) >= 3: # Minimum sample size for Shapiro-Wilk
        try:
            stat, p_value = shapiro(series_clean)
            print(f"\n{name}:")
            print(f"  Shapiro-Wilk Statistic: {stat:.4f}")
            print(f"  P-value: {p_value:.4f}")
            if p_value < 0.05:
                print("  -> Data likely NOT normally distributed (Reject H0)")
            else:
                print("  -> Data appears reasonably normal (Fail to reject H0)")
        except Exception as e:
            print(f"\n{name}: Error performing Shapiro-Wilk test - {e}")
            print("  (This might happen if data is constant or other issues)")
    else:
        print(f"\n{name}: Insufficient data ({len(series_clean)} observations after dropping NaNs) for Shapiro-Wilk test.")
```

```python
diq_file_names = [
    "DIQ_E.xpt",
    "DIQ_F.xpt",
    "DIQ_G.xpt",
    "DIQ_H.xpt",
    "DIQ_I.xpt",
    "DIQ_J.xpt"
]

diq_dfs = []

print("Loading and selecting Diabetes Questionnaire data:")
# Use the previously defined read_nhanes_xpt helper function
for file_name in diq_file_names:
    # Construct the URL dynamically based on the file name and NHANES_CYCLES structure
    url_year = None
    file_prefix = None
    for display_year_range, cycle_info in NHANES_CYCLES.items():
        if file_name.endswith(f"_{cycle_info['file_prefix']}.XPT") or file_name.endswith(f"_{cycle_info['file_prefix']}.xpt"):
            url_year = cycle_info['url_year']
            file_prefix = cycle_info['file_prefix']
            break

    if url_year and file_prefix:
        url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/{file_name}"
        df = read_nhanes_xpt(url, ['SEQN', 'DIQ010']) # Select SEQN and DIQ010
        if df is not None and not df.empty:
            print(f" Loaded and selected columns from: {file_name}")
            diq_dfs.append(df)
    else:
        print(f" Could not determine URL components for file: {file_name}")

df_diq = pd.concat(diq_dfs, ignore_index=True)
print(" Combined Diabetes Questionnaire shape:", df_diq.shape)

print("\nFirst few rows of the combined DIQ data:")
display(df_diq.head())

# Ensure df_combined_all is available
if 'df_combined_all' in locals():
    # Merge df_combined_all with df_diq
    # Use indicator=True to check merge results if needed for debugging, but removing for cleaner merge
    df_combined_all = pd.merge(df_combined_all, df_diq[['SEQN', 'DIQ010']], on='SEQN', how='left', suffixes=('_x', '_y'))
    print(" Merged with Diabetes Questionnaire data. Current shape:", df_combined_all.shape)

    # Determine the correct column name for DIQ010 after the merge
    # Assuming the latest merge adds the suffix _y
    diq_col_name = 'DIQ010_y'
    # Fallback if the column name is simply DIQ010 (e.g., if this is the first merge)
    if 'DIQ010' in df_combined_all.columns and 'DIQ010_y' not in df_combined_all.columns:
         diq_col_name = 'DIQ010'
    # If both _x and _y exist, and _y is from the latest merge, we'll use _y.
    # If for some reason only _x exists, use _x as a fallback.
    elif 'DIQ010_x' in df_combined_all.columns and 'DIQ010_y' not in df_combined_all.columns:
         diq_col_name = 'DIQ010_x'


    if diq_col_name and diq_col_name in df_combined_all.columns:
        print(f"\nUsing column '{diq_col_name}' for diabetes diagnosis count.")
        # Count participants diagnosed with diabetes (code 1.0 for 'Yes')
        # Referencing NHANES documentation, DIQ010 = 1 indicates 'Yes' to doctor diagnosis.
        # Handle potential NaN or other values in the DIQ010 column
        diagnosed_diabetes_count = df_combined_all[df_combined_all[diq_col_name] == 1.0].shape[0]

        # Calculate the total number of participants in the merged dataframe
        total_participants = df_combined_all.shape[0]

        # Calculate the percentage
        percentage_diagnosed = (diagnosed_diabetes_count / total_participants) * 100 if total_participants > 0 else 0

        print(f"\nPercentage of participants diagnosed with Type 2 Diabetes: {percentage_diagnosed:.2f}%")
    else:
        print("\n Could not find a valid 'DIQ010' column (or variants) in the DataFrame after merge.")


else:
    print(" 'df_combined_all' DataFrame not found. Please ensure preceding cells are run correctly.")
```

### **6.3 Calculate high-risk biomarker percentages**



```python
# Ensure df_combined_all is available
if 'df_combined_all' in locals() and not df_combined_all.empty:

    # 1. Define high-risk/abnormal thresholds
    # These are common clinical thresholds, but may vary slightly depending on guidelines.
    # Fasting Glucose: >= 100 mg/dL (Prediabetes) or >= 126 mg/dL (Diabetes)
    # For this task, let's use >= 100 mg/dL as "high-risk or abnormal range" as per the prompt's phrasing.
    glucose_threshold = 100 # mg/dL

    # HbA1c: >= 5.7% (Prediabetes) or >= 6.5% (Diabetes)
    # Let's use >= 5.7% as "high-risk or abnormal range".
    hba1c_threshold = 5.7 # %

    # Triglycerides: >= 150 mg/dL (High)
    triglycerides_threshold = 150 # mg/dL

    # Total number of participants in the relevant age group (from df_combined_all)
    total_participants = df_combined_all.shape[0]

    print(f"\nTotal participants in the dataset: {total_participants}")
    print("\n--- Percentage of Participants in High-Risk/Abnormal Biomarker Ranges ---")

    # 2. & 3. Create masks and calculate counts for each biomarker
    # Fasting Glucose
    # Filter out missing values before applying the threshold
    glucose_high_risk_mask = df_combined_all['LBXGLU'].dropna() >= glucose_threshold
    glucose_high_risk_count = glucose_high_risk_mask.sum()

    # HbA1c
    # Filter out missing values before applying the threshold
    hba1c_high_risk_mask = df_combined_all['LBXGH'].dropna() >= hba1c_threshold
    hba1c_high_risk_count = hba1c_high_risk_mask.sum()

    # Triglycerides
    # Filter out missing values before applying the threshold
    triglycerides_high_risk_mask = df_combined_all['LBXTR'].dropna() >= triglycerides_threshold
    triglycerides_high_risk_count = triglycerides_high_risk_mask.sum()

    # 4. Calculate percentages
    # Calculate percentage based on the number of participants with *valid* data for that biomarker
    # This is more informative than percentage of the total sample if many values are missing.
    # Total participants with valid glucose data
    total_valid_glucose = df_combined_all['LBXGLU'].dropna().shape[0]
    percentage_glucose_high_risk = (glucose_high_risk_count / total_valid_glucose) * 100 if total_valid_glucose > 0 else 0

    # Total participants with valid HbA1c data
    total_valid_hba1c = df_combined_all['LBXGH'].dropna().shape[0]
    percentage_hba1c_high_risk = (hba1c_high_risk_count / total_valid_hba1c) * 100 if total_valid_hba1c > 0 else 0

    # Total participants with valid Triglycerides data
    total_valid_triglycerides = df_combined_all['LBXTR'].dropna().shape[0]
    percentage_triglycerides_high_risk = (triglycerides_high_risk_count / total_valid_triglycerides) * 100 if total_valid_triglycerides > 0 else 0


    # 5. Print the results
    print(f"Fasting Glucose (>= {glucose_threshold} mg/dL):")
    print(f"  Count: {glucose_high_risk_count}")
    print(f"  Percentage (of {total_valid_glucose} valid entries): {percentage_glucose_high_risk:.2f}%")

    print(f"\nHbA1c (>= {hba1c_threshold} %):")
    print(f"  Count: {hba1c_high_risk_count}")
    print(f"  Percentage (of {total_valid_hba1c} valid entries): {percentage_hba1c_high_risk:.2f}%")


    print(f"\nTriglycerides (>= {triglycerides_threshold} mg/dL):")
    print(f"  Count: {triglycerides_high_risk_count}")
    print(f"  Percentage (of {total_valid_triglycerides} valid entries): {percentage_triglycerides_high_risk:.2f}%")


else:
    print(" 'df_combined_all' DataFrame not found or is empty. Please ensure preceding cells are run correctly.")
```

### **6.4 Calculate medication usage percentage**



```python
# Define file names for medication data (RXQ_RX) across cycles E-J
rxq_rx_file_names = [
    "RXQ_RX_E.xpt",
    "RXQ_RX_F.xpt",
    "RXQ_RX_G.xpt",
    "RXQ_RX_H.xpt",
    "RXQ_RX_I.xpt",
    "RXQ_RX_J.xpt"
]

rxq_rx_dfs = []

print("Loading and processing Medication data:")
# Use the previously defined read_nhanes_xpt helper function
for file_name in rxq_rx_file_names:
    # Construct the URL dynamically based on the file name and NHANES_CYCLES structure
    url_year = None
    file_prefix = None
    for display_year_range, cycle_info in NHANES_CYCLES.items():
        if file_name.endswith(f"_{cycle_info['file_prefix']}.XPT") or file_name.endswith(f"_{cycle_info['file_prefix']}.xpt"):
            url_year = cycle_info['url_year']
            file_prefix = cycle_info['file_prefix']
            break

    if url_year and file_prefix:
        url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/{file_name}"
        # Relevant columns in RXQ_RX are typically SEQN (Participant ID) and RXDDRUG (Generic Drug Name) or RXDDRGID (Generic Drug Code)
        # We need to identify medications related to Type 2 Diabetes.
        # A common approach is to look for specific drug names or classes.
        # Let's load SEQN and the drug name/code columns and then filter.
        # Checking documentation, RXDDRUG and RXDDRGID seem relevant.
        df = read_nhanes_xpt(url, ['SEQN', 'RXDDRUG', 'RXDDRGID'])
        if df is not None and not df.empty:
            print(f" Loaded and selected columns from: {file_name}")
            rxq_rx_dfs.append(df)
    else:
        print(f" Could not determine URL components for file: {file_name}")

df_rxq_rx = pd.concat(rxq_rx_dfs, ignore_index=True)
print(" Combined Medication data shape:", df_rxq_rx.shape)

# Decode the 'RXDDRUG' column from bytes to string (UTF-8)
if 'RXDDRUG' in df_rxq_rx.columns:
    # Apply decoding only if the column contains byte strings
    if df_rxq_rx['RXDDRUG'].dtype == 'object': # Check if dtype is object (can hold mixed types including bytes)
        # Check if at least one value is bytes before applying decode
        if any(isinstance(x, bytes) for x in df_rxq_rx['RXDDRUG'].dropna().head()): # Sample a few non-NaN to check
             df_rxq_rx['RXDDRUG'] = df_rxq_rx['RXDDRUG'].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
             print(" Decoded 'RXDDRUG' column from bytes to string.")
        else:
             print(" 'RXDDRUG' column does not appear to contain byte strings. Skipping decoding.")
    else:
         print(f" 'RXDDRUG' column has dtype {df_rxq_rx['RXDDRUG'].dtype}. Skipping decoding.")


print("\nFirst few rows of the combined Medication data after processing:")
display(df_rxq_rx.head())

# Ensure df_combined_all is available from previous steps
if 'df_combined_all' in locals() and not df_combined_all.empty:
    # Merge the combined medication data with the main dataset
    # We only need to know *if* a person is taking any diabetes medication,
    # so we can simplify df_rxq_rx to a list of unique SEQN before merging.
    # This prevents inflating the main dataframe size if a person takes multiple medications.

    # Identify participants taking diabetes medication
    # This requires a list of common Type 2 Diabetes medications.
    # Common classes/drugs include: Metformin, Sulfonylureas, DPP-4 inhibitors,
    # SGLT2 inhibitors, GLP-1 receptor agonists, Insulin.
    # We'll need to look for these (or their generic names/codes) in RXDDRUG or RXDDRGID.

    # Example list of generic diabetes drug names (case-insensitive) - this is not exhaustive
    diabetes_med_keywords = [
        'METFORMIN', 'GLIPIZIDE', 'GLYBURIDE', 'GLIMEPIRIDE', # Biguanides, Sulfonylureas
        'SITAGLIPTIN', 'SAXAGLIPTIN', 'LINAGLIPTIN', # DPP-4 inhibitors
        'CANAGLIFLOZIN', 'DAPAGLIFLOZIN', 'EMPAGLIFLOZIN', # SGLT2 inhibitors
        'EXENATIDE', 'LIRAGLUTIDE', 'DULAGLUTIDE', # GLP-1 receptor agonists
        'INSULIN', # Insulin
        'PIOGLITAZONE', 'ROSIGLITAZONE' # Thiazolidinediones
        # Add more as needed based on specific drug names in the data
    ]

    # Filter df_rxq_rx for rows containing diabetes medication keywords in RXDDRUG
    # Need to handle potential NaNs in RXDDRUG
    df_diabetes_meds = pd.DataFrame() # Initialize empty DataFrame

    if 'RXDDRUG' in df_rxq_rx.columns:
        # Ensure the column is string type before using .str accessor
        df_rxq_rx['RXDDRUG'] = df_rxq_rx['RXDDRUG'].astype(str)
        # Now apply the filter
        df_diabetes_meds = df_rxq_rx[
            df_rxq_rx['RXDDRUG'].str.contains('|'.join(diabetes_med_keywords), case=False, na=False)
        ].copy() # Use .copy() to avoid SettingWithCopyWarning
        print(f" Found {df_diabetes_meds.shape[0]} medication entries potentially related to diabetes.")
    else:
         print(" RXDDRUG column not found in medication data. Cannot filter for diabetes medications.")


    # Get unique participant IDs who are taking diabetes medication
    participants_on_diabetes_meds = df_diabetes_meds['SEQN'].unique()
    print(f" Identified {len(participants_on_diabetes_meds)} unique participants potentially on diabetes medication.")


    # Create a new column in df_combined_all indicating if the participant is on diabetes medication
    df_combined_all['On_Diabetes_Medication'] = df_combined_all['SEQN'].isin(participants_on_diabetes_meds)

    print("\nFirst few rows of df_combined_all with 'On_Diabetes_Medication' column:")
    display(df_combined_all.head())

    # Count participants on diabetes medication in the filtered df_combined_all
    on_medication_count = df_combined_all['On_Diabetes_Medication'].sum()

    # Calculate the total number of participants in df_combined_all
    total_participants_combined = df_combined_all.shape[0]

    # Calculate the percentage
    percentage_on_medication = (on_medication_count / total_participants_combined) * 100 if total_participants_combined > 0 else 0

    print(f"\nPercentage of participants taking medication related to Type 2 Diabetes: {percentage_on_medication:.2f}%")

else:
    print(" 'df_combined_all' DataFrame not found or is empty. Cannot proceed with merging and calculation.")
```
