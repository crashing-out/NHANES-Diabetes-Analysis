```python
# Start by importing necessary function libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm # Import the statsmodels library

# --- Data Loading ---
# URLs for the required NHANES 2017-2018 data files
demo_url = "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/DEMO_J.xpt" # Age and Gender
bmx_url = "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/BMX_J.xpt" # BMI
vid_url = "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/VID_J.xpt" # Serum Vitamin D (LBXVIDMS)
dbq_url = "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/DBQ_J.xpt" # Fast Food/Pizza Meals (DBD900)
slq_url = "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/SLQ_J.xpt" # Sleep Hours (SLD012)

print("Loading dataframes...")
demo = pd.read_sas(demo_url)
bmx = pd.read_sas(bmx_url)
vid = pd.read_sas(vid_url)
dbq = pd.read_sas(dbq_url)
slq = pd.read_sas(slq_url)
print("Dataframes loaded.")

# --- Data Merging ---
# Merge the dataframes step by step on 'SEQN' (Sequence Number, unique participant identifier)
# This ensures all relevant data for each participant is combined.
print("Merging dataframes...")
merged_df = demo.merge(bmx, on="SEQN", how="inner") \
                .merge(vid, on="SEQN", how="inner") \
                .merge(dbq, on="SEQN", how="inner") \
                .merge(slq, on="SEQN", how="inner")
print("Dataframes merged successfully.")

# Display the first few rows and columns of the merged dataframe to verify
print("\nMerged Dataframe Head:")
display(merged_df.head())
print("\nMerged Dataframe Columns:")
print(merged_df.columns)

# --- Data Extraction and Initial Cleaning ---
# Create an extracted dataframe containing only the variables of interest:
# RIDAGEYR - Age in years at screening
# RIAGENDR - Gender (1=Male, 2=Female)
# BMXBMI - Body Mass Index
# DBD900 - Number of meals from fast food or pizza place in the last 30 days
# SLD012 - How much sleep do you usually get at night (hours)
# LBXVIDMS - Serum Vitamin D (25OHD2 + 25OHD3) - This is our dependent variable (biomarker)
df = merged_df[["RIDAGEYR", "RIAGENDR", "BMXBMI", "DBD900", "SLD012", "LBXVIDMS"]].copy()

# Drop rows with any missing values in the selected columns
# This ensures that our analysis is based on complete cases for all variables.
print("\nInitial number of rows before dropping NaNs:", len(df))
df.dropna(inplace=True)
print("Number of rows after dropping NaNs:", len(df))

# Check for missing values after dropping (should all be 0)
print("\nMissing values after initial dropna:")
print(df.isnull().sum())

# --- Data Cleaning and Filtering based on requirements ---
print("\nApplying specific data filters...")

# Filter 1: Age (RIDAGEYR) - Valid ages <= 80
df = df[df["RIDAGEYR"] <= 80]
print(f"Rows after filtering RIDAGEYR <= 80: {len(df)}")

# Filter 2: Gender (RIAGENDR) - Exclude 7 (Refused) and 9 (Don't know)
# RIAGENDR typically uses 1 (Male) and 2 (Female). If 7 or 9 are present, remove them.
df = df[~df["RIAGENDR"].isin([7, 9])]
print(f"Rows after filtering RIAGENDR (excluding 7, 9): {len(df)}")

# Filter 3: Number of meals from fast food or pizza place (DBD900) - Valid number of meals <= 21
df = df[df["DBD900"] <= 21]
print(f"Rows after filtering DBD900 <= 21: {len(df)}")

# Filter 4: Sleep hours (SLD012) - Valid number of meals <= 24
df = df[df["SLD012"] <= 24]
print(f"Rows after filtering SLD012 <= 24: {len(df)}")

# Display the cleaned dataframe head
print("\nCleaned Dataframe Head after all general filters:")
display(df.head())
print(f"Final number of rows after all general filters: {len(df)}")

# --- Regression Analysis 1: Using the raw data (after general filters) ---
print("\n--- Performing Multilinear Regression (Model 1: Full Filtered Data) ---")

# Define the independent variables (X) and the dependent variable (Y)
# X includes Age, Gender, BMI, Fast Food Meals, and Sleep Hours
X1 = df[["RIDAGEYR", "RIAGENDR", "BMXBMI", "DBD900", "SLD012"]]
Y1 = df["LBXVIDMS"] # Dependent variable: Serum Vitamin D

# Add a constant to the independent variables for the regression model
# This represents the intercept of the regression line.
X1 = sm.add_constant(X1)

# Fit the Ordinary Least Squares (OLS) model
model1 = sm.OLS(Y1, X1).fit()

# Print the summary of the regression model
print("\nModel 1: Regression Summary (Full Filtered Data)")
print(model1.summary())

# --- Regression Analysis 2: Using a cut-off of BMI <= 50 ---
print("\n--- Performing Multilinear Regression (Model 2: BMI <= 50) ---")

# Create a new dataframe by filtering the already cleaned 'df' for BMXBMI <= 50
df_bmi_filtered = df[df["BMXBMI"] <= 50].copy()
print(f"Number of rows for Model 2 (BMXBMI <= 50): {len(df_bmi_filtered)}")

# Define the independent variables (X) and the dependent variable (Y) for this filtered dataset
X2 = df_bmi_filtered[["RIDAGEYR", "RIAGENDR", "BMXBMI", "DBD900", "SLD012"]]
Y2 = df_bmi_filtered["LBXVIDMS"] # Dependent variable: Serum Vitamin D

# Add a constant to the independent variables for the regression model
X2 = sm.add_constant(X2)

# Fit the Ordinary Least Squares (OLS) model
model2 = sm.OLS(Y2, X2).fit()

# Print the summary of the regression model
print("\nModel 2: Regression Summary (BMXBMI <= 50)")
print(model2.summary())
```

```python
import pandas as pd
import numpy as np
import requests # Import requests for robust URL checking
from io import BytesIO # Import BytesIO to read content directly from requests
from tqdm.notebook import tqdm # For progress bars in notebooks/interactive environments

# --- Configuration ---
# Define years and cycles for consistent URL construction
# Note: For 2017-2018 (J) and 2018-2020 (J), the 'year' in the URL is '2017'.
# The actual data cycle code for 2018-2020 is also 'J'.
# NHANES data is released in 2-year cycles. The "2020" you have is
# effectively the 2017-2018 data due to COVID-19 related changes and
# continuous release. For consistency with how CDC names files,
# we'll map years to the directory year and cycle code correctly.

# NHANES URL mapping: {display_year_range: {'url_year': 'year_in_url', 'file_prefix': 'cycle_code'}}
# I'm using the standard 2-year cycle notation which is clearer for NHANES.
NHANES_CYCLES = {
    "2007-2008": {'url_year': '2007', 'file_prefix': 'E'},
    "2009-2010": {'url_year': '2009', 'file_prefix': 'F'},
    "2011-2012": {'url_year': '2011', 'file_prefix': 'G'},
    "2013-2014": {'url_year': '2013', 'file_prefix': 'H'},
    "2015-2016": {'url_year': '2015', 'file_prefix': 'I'},
    "2017-2018": {'url_year': '2017', 'file_prefix': 'J'},
    # The 2018-2020 (post-COVID) data often references 2017 files for consistency
    # as no new full cycle data was released. It uses the 'J' prefix.
    # If your analysis specifically needs to distinguish, you'd need to consult
    # specific NHANES documentation for this specific data release.
    # For strict 2017-2020 data, NHANES usually calls it "Continuous NHANES 2017-March 2020 Pre-Pandemic File"
    # and it uses the 'J' cycle code for files.
    "2017-2020": {'url_year': '2017', 'file_prefix': 'J'} # Using '2017-2020' as a key for this range.
}


# --- Helper Function for Robust URL Reading ---
def read_nhanes_xpt(url, cols_to_select=None):
    """
    Reads an XPT file directly from a URL using requests and pandas.read_sas.
    Handles potential file not found errors and selects specified columns.
    Explicitly sets format='xport'.
    """
    try:
        response = requests.get(url)
        response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
        # Use BytesIO to read content directly from the response
        df = pd.read_sas(BytesIO(response.content), format='xport') # Explicitly set format
        if cols_to_select:
            # Select only columns that exist in the DataFrame
            existing_cols = [col for col in cols_to_select if col in df.columns]
            if len(existing_cols) < len(cols_to_select):
                print(f"Warning: Some requested columns not found in {url.split('/')[-1]}: {set(cols_to_select) - set(existing_cols)}")
            df = df[existing_cols]
        return df
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {url}: {e}")
        return None
    except Exception as e:
        print(f"Error reading XPT from {url}: {e}")
        return None

# --- Data Processing Functions ---

# Removed process_liver_data function


def process_demographics(df_demo_raw):
    if df_demo_raw is None or df_demo_raw.empty:
        return pd.DataFrame()
    relevant_cols = ['SEQN', 'RIDAGEYR', 'RIAGENDR', 'RIDRETH1']
    df_demo = df_demo_raw[[col for col in relevant_cols if col in df_demo_raw.columns]].copy()
    for col in ['RIDAGEYR', 'RIAGENDR', 'RIDRETH1']:
        if col in df_demo.columns:
            df_demo[col] = pd.to_numeric(df_demo[col], errors='coerce')
    return df_demo

def process_bmi_data(df_bmi_raw):
    if df_bmi_raw is None or df_bmi_raw.empty:
        return pd.DataFrame()
    relevant_cols = ['SEQN', 'BMXBMI', 'BMXWAIST', 'BMXHIP'] # Adding waist and hip for completeness
    df_bmi = df_bmi_raw[[col for col in relevant_cols if col in df_bmi_raw.columns]].copy()
    for col in ['BMXBMI', 'BMXWAIST', 'BMXHIP']:
        if col in df_bmi.columns:
            df_bmi[col] = pd.to_numeric(df_bmi[col], errors='coerce')
    return df_bmi

def process_sleep_data(df_sleep_raw):
    if df_sleep_raw is None or df_sleep_raw.empty:
        return pd.DataFrame()

    df_sleep = df_sleep_raw.copy()
    # Prioritize SLD012/SLD013 as discussed for average sleep duration
    # SLD010H was present in older cycles, but SLD012/SLD013 are more common in recent ones.
    # For this modified script, we'll favor the SLD012/SLD013 average approach.
    sleep_cols_to_check = ['SLD012', 'SLD013', 'SLD010H'] # Order of preference

    # Convert to numeric, handle NHANES missing codes
    for col in sleep_cols_to_check:
        if col in df_sleep.columns:
            df_sleep[col] = pd.to_numeric(df_sleep[col], errors='coerce').replace({7: np.nan, 9: np.nan, 77: np.nan, 99: np.nan})

    df_sleep['Avg_Sleep_Hours'] = np.nan

    if 'SLD012' in df_sleep.columns and 'SLD013' in df_sleep.columns:
        # Calculate average of weekday and weekend sleep
        df_sleep['Avg_Sleep_Hours'] = df_sleep[['SLD012', 'SLD013']].mean(axis=1)
    elif 'SLD012' in df_sleep.columns: # Fallback if only weekday sleep is available/valid
        df_sleep['Avg_Sleep_Hours'] = df_sleep['SLD012']
    elif 'SLD010H' in df_sleep.columns: # Older cycle alternative
        df_sleep['Avg_Sleep_Hours'] = df_sleep['SLD010H']
    else:
        print(f"Warning: No valid sleep duration variable found in this cycle's SLQ data.")

    return df_sleep[['SEQN', 'Avg_Sleep_Hours']].copy()

# Removed process_alcohol_data function


def process_physical_activity_data(df_pa_raw):
    """
    Processes raw PAQ data to calculate weekly vigorous recreational physical activity.
    """
    if df_pa_raw is None or df_pa_raw.empty:
        return pd.DataFrame()

    df_pa = df_pa_raw.copy()
    pa_cols = ['SEQN', 'PAQ650', 'PAQ655', 'PAD660'] # Screener, Days, Minutes

    for col in pa_cols:
        if col in df_pa.columns:
            df_pa[col] = pd.to_numeric(df_pa[col], errors='coerce')
            # Replace common NHANES missing codes
            if col == 'PAQ650': df_pa[col] = df_pa[col].replace({7: np.nan, 9: np.nan})
            elif col == 'PAQ655': df_pa[col] = df_pa[col].replace({77: np.nan, 99: np.nan})
            elif col == 'PAD660': df_pa[col] = df_pa[col].replace({7777: np.nan, 9999: np.nan})
        else: # If a critical column is missing, fill with NaN and print warning
            df_pa[col] = np.nan
            print(f"Warning: Missing PAQ column '{col}' for PA processing. May affect results.")


    # Initialize the new calculated column
    df_pa['Vigorous_Rec_Min_Week'] = np.nan # Start with NaN to distinguish from 0

    # Condition for valid vigorous activity data (PAQ650=Yes, and days/minutes are not NaN)
    # Ensure that PAQ655 and PAD660 columns exist before using them
    if 'PAQ655' in df_pa.columns and 'PAD660' in df_pa.columns:
        valid_vig_pa_cond = (df_pa['PAQ650'] == 1) & \
                            (~df_pa['PAQ655'].isna()) & \
                            (~df_pa['PAD660'].isna())

        # Perform the calculation for valid rows
        df_pa.loc[valid_vig_pa_cond, 'Vigorous_Rec_Min_Week'] = \
            df_pa['PAQ655'] * df_pa['PAD660']

    # Set to 0.0 for those who explicitly said 'No' to vigorous activity (PAQ650 == 2)
    if 'PAQ650' in df_pa.columns:
         df_pa.loc[df_pa['PAQ650'] == 2, 'Vigorous_Rec_Min_Week'] = 0.0


    return df_pa[['SEQN', 'Vigorous_Rec_Min_Week']].copy()

def process_diet_data(df_dr1tot_raw, df_dr2tot_raw):
    """
    Processes raw DR1TOT and DR2TOT data to calculate average total sugar intake.
    """
    if (df_dr1tot_raw is None or df_dr1tot_raw.empty) and \
       (df_dr2tot_raw is None or df_dr2tot_raw.empty):
        print("DR1TOT and DR2TOT DataFrames are empty or not loaded, skipping diet processing.")
        return pd.DataFrame()

    # Create copies to avoid SettingWithCopyWarning
    dr1 = df_dr1tot_raw[['SEQN', 'DR1TSUGR', 'DR1DRSTZ']].copy() if df_dr1tot_raw is not None and not df_dr1tot_raw.empty else pd.DataFrame()
    dr2 = df_dr2tot_raw[['SEQN', 'DR2TSUGR', 'DR2DRSTZ']].copy() if df_dr2tot_raw is not None and not df_dr2tot_raw.empty else pd.DataFrame()


    # Rename columns for merging
    if not dr1.empty: dr1.rename(columns={'DR1TSUGR': 'Total_Sugar_Day1', 'DR1DRSTZ': 'Recall_Status_Day1'}, inplace=True)
    if not dr2.empty: dr2.rename(columns={'DR2TSUGR': 'Total_Sugar_Day2', 'DR2DRSTZ': 'Recall_Status_Day2'}, inplace=True)


    # Merge the two dataframes. Use 'outer' to keep all SEQN from either day.
    if not dr1.empty and not dr2.empty:
        diet_df = pd.merge(dr1, dr2, on='SEQN', how='outer')
    elif not dr1.empty:
        diet_df = dr1
    elif not dr2.empty:
        diet_df = dr2
    else:
        return pd.DataFrame({'SEQN': [], 'Avg_Total_Sugar_g': []}) # Return empty if neither loaded

    # Convert sugar values and recall status to numeric, handle missing codes
    for col in ['Total_Sugar_Day1', 'Total_Sugar_Day2', 'Recall_Status_Day1', 'Recall_Status_Day2']:
        if col in diet_df.columns:
            diet_df[col] = pd.to_numeric(diet_df[col], errors='coerce')
            # Assuming 1.0 means reliable. Non-1.0 status or missing should not be used for average.
            if 'Recall_Status' in col:
                diet_df.loc[diet_df[col].notna() & (diet_df[col] != 1), col] = np.nan # Only set to NaN if not 1 and not already NaN


    # Calculate average total sugar intake
    diet_df['Avg_Total_Sugar_g'] = np.nan

    # Case 1: Both days are reliable
    cond_both_reliable = (~diet_df['Total_Sugar_Day1'].isna()) & (diet_df['Recall_Status_Day1'] == 1.0) & \
                         (~diet_df['Total_Sugar_Day2'].isna()) & (diet_df['Recall_Status_Day2'] == 1.0)
    if 'Total_Sugar_Day1' in diet_df.columns and 'Total_Sugar_Day2' in diet_df.columns and 'Recall_Status_Day1' in diet_df.columns and 'Recall_Status_Day2' in diet_df.columns:
        diet_df.loc[cond_both_reliable, 'Avg_Total_Sugar_g'] = (diet_df['Total_Sugar_Day1'] + diet_df['Total_Sugar_Day2']) / 2


    # Case 2: Only Day 1 is reliable
    cond_day1_only = (~diet_df['Total_Sugar_Day1'].isna()) & (diet_df['Recall_Status_Day1'] == 1.0) & \
                     (diet_df['Total_Sugar_Day2'].isna() | (diet_df['Recall_Status_Day2'] != 1.0)) # Check if Day2 is NaN or not reliable
    if 'Total_Sugar_Day1' in diet_df.columns and 'Recall_Status_Day1' in diet_df.columns and 'Total_Sugar_Day2' in diet_df.columns and 'Recall_Status_Day2' in diet_df.columns:
         diet_df.loc[cond_day1_only, 'Avg_Total_Sugar_g'] = diet_df['Total_Sugar_Day1']


    # Case 3: Only Day 2 is reliable
    cond_day2_only = (diet_df['Total_Sugar_Day1'].isna() | (diet_df['Recall_Status_Day1'] != 1.0)) & \
                     (~diet_df['Total_Sugar_Day2'].isna()) & (diet_df['Recall_Status_Day2'] == 1.0) # Check if Day1 is NaN or not reliable
    if 'Total_Sugar_Day1' in diet_df.columns and 'Recall_Status_Day1' in diet_df.columns and 'Total_Sugar_Day2' in diet_df.columns and 'Recall_Status_Day2' in diet_df.columns:
        diet_df.loc[cond_day2_only, 'Avg_Total_Sugar_g'] = diet_df['Total_Sugar_Day2']


    return diet_df[['SEQN', 'Avg_Total_Sugar_g']].copy()


def process_smoking_data(df_smq_raw):
    """
    Processes raw SMQ data to create a combined smoking status variable.
    """
    if df_smq_raw is None or df_smq_raw.empty:
        return pd.DataFrame()

    df_smq = df_smq_raw.copy()
    relevant_cols = ['SEQN', 'SMQ020', 'SMQ040']

    for col in relevant_cols:
        if col in df_smq.columns:
            df_smq[col] = pd.to_numeric(df_smq[col], errors='coerce').replace({7: np.nan, 9: np.nan})
        else:
            df_smq[col] = np.nan # Add column if missing

    df_smq['Smoking_Status'] = pd.NA

    # Ensure columns exist before using them in conditions
    if 'SMQ020' in df_smq.columns:
        # Never Smoker: SMQ020 == 2
        df_smq.loc[df_smq['SMQ020'] == 2, 'Smoking_Status'] = 'Never Smoker'

        if 'SMQ040' in df_smq.columns:
             # Current Smoker: SMQ020 == 1 AND (SMQ040 == 1 or SMQ040 == 2)
            df_smq.loc[(df_smq['SMQ020'] == 1) & (df_smq['SMQ040'].isin([1, 2])), 'Smoking_Status'] = 'Current Smoker'

            # Former Smoker: SMQ020 == 1 AND SMQ040 == 3
            df_smq.loc[(df_smq['SMQ020'] == 1) & (df_smq['SMQ040'] == 3), 'Smoking_Status'] = 'Former Smoker'


    # Fill any remaining NA values (e.g., if SMQ020 was missing or conditions not met)
    df_smq['Smoking_Status'] = df_smq['Smoking_Status'].fillna('Unknown')

    return df_smq[['SEQN', 'Smoking_Status']].copy()

def process_family_history_data(df_diq_raw):
    """
    Processes raw DIQ data to create a family history of diabetes variable.
    """
    if df_diq_raw is None or df_diq_raw.empty:
        return pd.DataFrame()

    df_diq = df_diq_raw.copy()
    # Relevant DIQ columns, handling potential absence in older cycles
    diq_cols = ['SEQN', 'DIQ010', 'DIQ050', 'DIQ070', 'DIQ080A', 'DIQ080B', 'DIQ080C', 'DIQ080D', 'DIQ080E', 'DIQ080F', 'DIQ080G']
    for col in diq_cols:
        if col in df_diq.columns:
            df_diq[col] = pd.to_numeric(df_diq[col], errors='coerce').replace({7: np.nan, 9: np.nan, 77: np.nan, 99: np.nan})
        else:
            df_diq[col] = np.nan # Add column if missing for consistency

    df_diq['Family_History_Diabetes'] = 'Unknown/Missing' # Default value
    df_diq['Self_Reported_Diabetes_Status'] = df_diq['DIQ010'].map({1: 'Yes', 2: 'No', 3: 'Borderline'}).fillna('Unknown') # Also process DIQ010 here


    # Step 1: No Family History (DIQ050 == 2)
    if 'DIQ050' in df_diq.columns:
        df_diq.loc[df_diq['DIQ050'] == 2, 'Family_History_Diabetes'] = 'No Family History'

        # Step 2: Identify "Any Family History" based on DIQ050 == 1
        has_any_fh = (df_diq['DIQ050'] == 1)

        # Step 3: Check for First-Degree Relatives (Mother, Father, Siblings, Children)
        first_degree_cols = ['DIQ080A', 'DIQ080B', 'DIQ080C', 'DIQ080D']
        # Ensure these columns exist before checking
        first_degree_exist = [col for col in first_degree_cols if col in df_diq.columns]

        if first_degree_exist:
            df_diq['has_first_degree'] = df_diq[first_degree_exist].eq(1.0).any(axis=1)
            df_diq.loc[has_any_fh & df_diq['has_first_degree'], 'Family_History_Diabetes'] = 'First-Degree Relative'
        else:
            print(f"Warning: First-degree relative detail columns not found in DIQ for some cycles.")
            df_diq['has_first_degree'] = False # Default to False if cols don't exist

        # Step 4: Check for Second-Degree Relatives (Grandparents, Aunts/Uncles, Other)
        second_degree_cols = ['DIQ080E', 'DIQ080F', 'DIQ080G']
        second_degree_exist = [col for col in second_degree_cols if col in df_diq.columns]

        if second_degree_exist:
            df_diq['has_second_degree'] = df_diq[second_degree_exist].eq(1.0).any(axis=1)
            # Apply if they have *any* family history, no first-degree, AND have second-degree
            df_diq.loc[has_any_fh & ~df_diq['has_first_degree'] & df_diq['has_second_degree'], 'Family_History_Diabetes'] = 'Second-Degree Only'
        else:
            print(f"Warning: Second-degree relative detail columns not found in DIQ for some cycles.")
            df_diq['has_second_degree'] = False # Default to False if cols don't exist

        # Fallback for 'Any Family History' if detailed relatives couldn't be determined (e.g., missing DIQ080 info)
        df_diq.loc[has_any_fh & df_diq['Family_History_Diabetes'].isin(['Unknown/Missing']), 'Family_History_Diabetes'] = 'Any Family History (Details Unknown)'

        # Clean up temporary columns
        df_diq.drop(columns=['has_first_degree', 'has_second_degree'], errors='ignore', inplace=True)


    return df_diq[['SEQN', 'Family_History_Diabetes', 'Self_Reported_Diabetes_Status']].copy() # Return both family history and self-reported status



# --- Main Data Loading and Merging Process ---

all_df_list = []

# Map the cycle keys from NHANES_CYCLES for iteration
# This ensures we loop through the defined 2-year cycles including the 2017-2020 special case.
for display_year_range, cycle_info in tqdm(NHANES_CYCLES.items(), desc="Processing NHANES Cycles"):
    url_year = cycle_info['url_year']
    file_prefix = cycle_info['file_prefix']

    print(f"\n--- Loading data for {display_year_range} (URL Year: {url_year}, File Prefix: {file_prefix}) ---")

    try:
        # Define URLs for required files for the current cycle
        # Corrected URLs to use wwwn.cdc.gov and dynamic year/prefix
        file_urls = {
            # Removed BIOPRO and ALQ
            'DEMO':   f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/DEMO_{file_prefix}.XPT",
            'BMX':    f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/BMX_{file_prefix}.XPT",
            'SLQ':    f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/SLQ_{file_prefix}.XPT",
            'PAQ':    f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/PAQ_{file_prefix}.XPT",
            'DR1TOT': f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/DR1TOT_{file_prefix}.XPT",
            'DR2TOT': f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{url_year}/DataFiles/DR2TOT_{file_prefix}.XPT",
            # Removed SMQ and DIQ
        }

        # Load raw data for the current cycle
        # Pass the specific columns expected for each file, for efficiency and error handling.
        # Use the read_nhanes_xpt helper function
        # Removed df_liver_raw and df_alcohol_raw loading
        df_demo_raw = read_nhanes_xpt(file_urls['DEMO'], ['SEQN', 'RIDAGEYR', 'RIAGENDR', 'RIDRETH1'])
        df_bmi_raw = read_nhanes_xpt(file_urls['BMX'], ['SEQN', 'BMXBMI', 'BMXWAIST', 'BMXHIP'])
        df_sleep_raw = read_nhanes_xpt(file_urls['SLQ'], ['SEQN', 'SLD010H', 'SLD012', 'SLD013']) # Include all possible sleep cols
        df_pa_raw = read_nhanes_xpt(file_urls['PAQ'], ['SEQN', 'PAQ650', 'PAQ655', 'PAD660'])
        df_dr1tot_raw = read_nhanes_xpt(file_urls['DR1TOT'], ['SEQN', 'DR1TSUGR', 'DR1DRSTZ'])
        df_dr2tot_raw = read_nhanes_xpt(file_urls['DR2TOT'], ['SEQN', 'DR2TSUGR', 'DR2DRSTZ'])
        # Removed df_smq_raw and df_diq_raw loading


        # Process data using dedicated functions
        # Removed df_liver and df_alcohol processing
        df_demo = process_demographics(df_demo_raw)
        df_bmi = process_bmi_data(df_bmi_raw)
        df_sleep = process_sleep_data(df_sleep_raw)
        df_pa = process_physical_activity_data(df_pa_raw)
        df_diet = process_diet_data(df_dr1tot_raw, df_dr2tot_raw)
        # Removed df_smoking and df_family_history processing


        # Merge all processed dataframes for the current cycle
        # Start with demographics as the base, but only if it's not empty
        if df_demo is not None and not df_demo.empty:
            current_cycle_df = df_demo.copy() # Create a copy to avoid modifying the original processed df

            # List of DataFrames to merge, exclude if empty or None
            # Removed df_liver, df_alcohol, df_smoking, and df_family_history from the list
            dfs_to_merge = [df_bmi, df_sleep, df_pa, df_diet]

            for df_to_merge in dfs_to_merge:
                if df_to_merge is not None and not df_to_merge.empty and 'SEQN' in df_to_merge.columns:
                    current_cycle_df = pd.merge(current_cycle_df, df_to_merge, on="SEQN", how="left")
                elif df_to_merge is not None and not df_to_merge.empty and 'SEQN' not in df_to_merge.columns:
                     print(f"Warning: Skipping merge for a dataframe missing 'SEQN' column in cycle {display_year_range}.")


            current_cycle_df["Cycle"] = display_year_range # Add the human-readable cycle label
            all_df_list.append(current_cycle_df)

            print(f"✅ Loaded and merged data for cycle: {display_year_range}")
        else:
             print(f"Skipping merge for cycle {display_year_range} as demographics data was not loaded.")


    except Exception as e:
        print(f"❌ Failed to process cycle {display_year_range} - Error: {e}")


# Combine all cycles into one final DataFrame
if all_df_list:
    df_final = pd.concat(all_df_list, ignore_index=True)
    print("\n--- Final Combined Dataset ---")
    print("✅ Combined Dataset Shape:", df_final.shape)
    print("Cycle Distribution:\n", df_final["Cycle"].value_counts(dropna=False))
    print("\nHead of Final Dataset:")
    print(df_final.head())

    # Check and print missing values for key new columns, only if they exist
    # Removed liver and alcohol related columns from this check
    key_cols_to_check = ['Vigorous_Rec_Min_Week', 'Avg_Total_Sugar_g', 'Avg_Sleep_Hours'] # Removed smoking and diabetes related columns
    existing_key_cols = [col for col in key_cols_to_check if col in df_final.columns]
    if existing_key_cols:
        print("\nMissing values in key new columns:")
        print(df_final[existing_key_cols].isnull().sum())
    else:
        print("\nNo key processed columns found in the final dataset.")


    # Example: Filter out participants with pre-existing diabetes if predicting incidence
    # Use Self_Reported_Diabetes_Status created in process_family_history_data
    # Removed filtering related to diabetes status

else:
    print("❌ No datasets were loaded successfully.")
```

### Group Comparison: Biomarker Levels by Gender

**Purpose**: This code block compares the distributions of continuous biomarker variables (Fasting Glucose, HbA1c, Triglyceride) between the two gender groups using the Mann-Whitney U test.

**Key variables**:
- `df_combined_all`: The DataFrame containing all merged and imputed data.
- `Gender`: Categorical variable representing gender (1 = Male, 2 = Female).
- `LBXGLU`: Fasting Glucose concentration (continuous).
- `LBXGH`: HbA1c concentration (continuous).
- `LBXTR`: Triglyceride concentration (continuous).

**Rationale**:
The research question involves understanding how different biomarkers affect HbA1c levels in individuals of both genders. Comparing biomarker levels between genders is a crucial initial step to see if there are significant differences that might influence subsequent analyses, such as regression modeling. The Mann-Whitney U test is chosen because the biomarker distributions are not normally distributed, making it a more robust choice than the independent samples t-test for comparing two independent groups.

**Results**:
The output will show the results of the Mann-Whitney U test for each biomarker when comparing males and females. This includes the test statistic and the p-value. A low p-value (typically < 0.05) indicates a statistically significant difference in the distribution of the biomarker between the two gender groups. A discussion of the results will interpret the findings in the context of gender differences in biomarker levels.

```python
from scipy.stats import mannwhitneyu

# Ensure df_combined_all is available and contains the necessary columns
if 'df_combined_all' in locals() and 'Gender' in df_combined_all.columns and \
   'LBXGLU' in df_combined_all.columns and 'LBXGH' in df_combined_all.columns and 'LBXTR' in df_combined_all.columns:

    # Define the biomarker columns and their names for output
    biomarker_cols = ['LBXGLU', 'LBXGH', 'LBXTR']
    biomarker_names = ['Fasting Glucose', 'HbA1c', 'Triglyceride']

    print("--- Mann-Whitney U Test for Biomarker Levels by Gender ---")
    print("Comparing distributions between Gender = 1.0 (Male) and Gender = 2.0 (Female)")
    print("Null Hypothesis (H0): The distribution of the biomarker is the same across both genders.")
    print("Alternative Hypothesis (H1): The distribution of the biomarker is different across genders.")
    print("Interpretation: A p-value < 0.05 leads to rejection of H0, suggesting a significant difference.")

    # Separate data by gender
    gender_male = df_combined_all[df_combined_all['Gender'] == 1.0]
    gender_female = df_combined_all[df_combined_all['Gender'] == 2.0]

    if gender_male.empty or gender_female.empty:
        print("\n❗ Warning: One or both gender groups are empty. Cannot perform Mann-Whitney U test.")
    else:
        for col, name in zip(biomarker_cols, biomarker_names):
            if col in df_combined_all.columns:
                # Extract biomarker data for each gender, dropping NaNs (already imputed, but good practice)
                male_data = gender_male[col].dropna()
                female_data = gender_female[col].dropna()

                if len(male_data) > 0 and len(female_data) > 0:
                    try:
                        stat, p_value = mannwhitneyu(male_data, female_data, alternative='two-sided')
                        print(f"\n{name}:")
                        print(f"  Mann-Whitney U Statistic: {stat:.4f}")
                        print(f"  P-value: {p_value:.4f}")
                        if p_value < 0.05:
                            print("  -> Statistically significant difference between genders (Reject H0)")
                        else:
                            print("  -> No statistically significant difference between genders (Fail to reject H0)")
                    except Exception as e:
                        print(f"\n{name}: Error performing Mann-Whitney U test - {e}")
                else:
                    print(f"\n{name}: Insufficient data in one or both gender groups after dropping NaNs. Cannot perform test.")
            else:
                print(f"\n❗ Warning: Biomarker column '{col}' not found in df_combined_all. Skipping test.")


    print("\nDiscussion of Results:")
    print("The Mann-Whitney U test results indicate whether there are statistically significant differences in the distributions of Fasting Glucose, HbA1c, and Triglyceride levels between males and females.")
    print("A significant p-value suggests that gender is associated with the level of that biomarker.")
    print("These findings are important for understanding potential gender-specific patterns in biomarker levels and for informing subsequent regression models, where gender may be included as a predictor.")

else:
    print("❗ 'df_combined_all' DataFrame or required columns ('Gender', 'LBXGLU', 'LBXGH', 'LBXTR') not found. Please ensure preceding cells are run correctly.")
```

### Group Comparison: Biomarker Levels by BMI Category

**Purpose**: This code block compares the distributions of continuous biomarker variables (Fasting Glucose, HbA1c, Triglyceride) across different BMI categories using the Kruskal-Wallis H test.

**Key variables**:
- `df_combined_all`: The DataFrame containing all merged and imputed data.
- `BMI_Category`: Categorical variable representing BMI categories ('Underweight', 'Healthy weight', 'Overweight', 'Obesity').
- `LBXGLU`: Fasting Glucose concentration (continuous).
- `LBXGH`: HbA1c concentration (continuous).
- `LBXTR`: Triglyceride concentration (continuous).

**Rationale**:
Understanding if biomarker levels differ significantly across BMI categories is important as BMI is a known risk factor for various health conditions, including diabetes. The Kruskal-Wallis H test is appropriate here because we are comparing a continuous variable across more than two independent groups (BMI categories) and the biomarker data is not normally distributed. If the Kruskal-Wallis test is significant, it indicates that at least one BMI category has a different biomarker distribution compared to others, warranting post-hoc analysis to pinpoint the specific differences.

**Results**:
The output will show the results of the Kruskal-Wallis H test for each biomarker when comparing the BMI categories. This includes the test statistic and the p-value. A low p-value (typically < 0.05) indicates a statistically significant difference in the distribution of the biomarker across the BMI categories. A discussion will interpret these findings and indicate if post-hoc analysis is needed.

```python
from scipy.stats import kruskal

# Ensure df_combined_all is available and contains the necessary columns
if 'df_combined_all' in locals() and 'BMI_Category' in df_combined_all.columns and \
   'LBXGLU' in df_combined_all.columns and 'LBXGH' in df_combined_all.columns and 'LBXTR' in df_combined_all.columns:

    # Define the biomarker columns and their names for output
    biomarker_cols = ['LBXGLU', 'LBXGH', 'LBXTR']
    biomarker_names = ['Fasting Glucose', 'HbA1c', 'Triglyceride']

    # Get unique BMI categories, dropping NaN if necessary for grouping
    bmi_categories = df_combined_all['BMI_Category'].dropna().unique()

    if len(bmi_categories) < 2:
        print("\n❗ Warning: Not enough unique BMI categories (need at least 2) to perform Kruskal-Wallis test after dropping NaNs.")
    else:
        print("--- Kruskal-Wallis H Test for Biomarker Levels by BMI Category ---")
        print("Null Hypothesis (H0): The distribution of the biomarker is the same across all BMI categories.")
        print("Alternative Hypothesis (H1): The distribution of the biomarker is different for at least one BMI category.")
        print("Interpretation: A p-value < 0.05 leads to rejection of H0, suggesting a significant difference.")

        for col, name in zip(biomarker_cols, biomarker_names):
            if col in df_combined_all.columns:
                # Prepare data for Kruskal-Wallis: list of arrays/series for each group
                data_by_category = [df_combined_all[df_combined_all['BMI_Category'] == category][col].dropna() for category in bmi_categories]

                # Filter out empty data arrays/series before the test
                data_by_category_cleaned = [data for data in data_by_category if len(data) > 0]
                categories_present = [bmi_categories[i] for i, data in enumerate(data_by_category) if len(data) > 0]


                if len(data_by_category_cleaned) >= 2: # Need at least two groups with data
                    try:
                        stat, p_value = kruskal(*data_by_category_cleaned)
                        print(f"\n{name}:")
                        print(f"  Kruskal-Wallis H Statistic: {stat:.4f}")
                        print(f"  P-value: {p_value:.4f}")
                        if p_value < 0.05:
                            print("  -> Statistically significant difference between BMI categories (Reject H0)")
                            print("     Post-hoc analysis (e.g., Dunn's test) may be needed to find which groups differ.")
                        else:
                            print("  -> No statistically significant difference between BMI categories (Fail to reject H0)")
                        print(f"  Categories included in test: {categories_present}")

                    except Exception as e:
                        print(f"\n{name}: Error performing Kruskal-Wallis test - {e}")
                else:
                    print(f"\n{name}: Insufficient data in BMI categories after dropping NaNs for Kruskal-Wallis test.")
                    print(f"  Categories with data: {categories_present}")


            else:
                print(f"\n❗ Warning: Biomarker column '{col}' not found in df_combined_all. Skipping test.")


    print("\nDiscussion of Results:")
    print("The Kruskal-Wallis H test results indicate whether there are statistically significant differences in the distributions of Fasting Glucose, HbA1c, and Triglyceride levels across the different BMI categories.")
    print("A significant p-value suggests that BMI category is associated with the level of that biomarker.")
    print("If a significant difference is found, further post-hoc tests are required to determine which specific BMI categories have significantly different biomarker levels.")

else:
    print("❗ 'df_combined_all' DataFrame or required columns ('BMI_Category', 'LBXGLU', 'LBXGH', 'LBXTR') not found. Please ensure preceding cells are run correctly.")
```

Multiple linear regression is a suitable method to assess how multiple predictors collectively influence a continuous outcome variable (HbA1c). This approach allows us to estimate the unique contribution of each predictor to the model while holding other predictors constant. It helps in understanding the complex interplay of factors associated with HbA1c levels, which is a key biomarker for diabetes management.

**Results**:
The output will display the summary of the multiple linear regression model. This summary typically includes:
- **R-squared and Adjusted R-squared**: Indicate the proportion of variance in HbA1c explained by the model.
- **F-statistic and p-value**: Test the overall significance of the model (whether at least one predictor is significantly associated with HbA1c).
- **Coefficients**: The estimated change in HbA1c for a one-unit increase in each predictor (holding others constant).
- **P-values for coefficients**: Indicate the statistical significance of each individual predictor.
- **Standard Errors**: Measure the precision of the coefficient estimates.

A discussion will interpret these results in the context of the research question, highlighting the significant predictors and the overall fit of the model.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd # Import pandas to create a DataFrame for the results

# Ensure df_combined_all is available and contains the necessary columns
if 'df_combined_all' in locals():

    # Define the dependent variable
    dependent_variable = 'LBXGH'

    # Define the independent variables (predictors)
    # Include relevant continuous and categorical variables.
    # Categorical variables need to be handled correctly in the formula (e.g., using C() or one-hot encoding).
    # Let's start with a selection of predictors based on common risk factors and our data.
    # We'll include Age, Gender, BMI, Smoking Status, Avg_Sleep_Hours, Vigorous_Rec_Min_Week, Avg_Total_Sugar_g, LBXGLU, LBXTR
    # For categorical variables like Gender and Smoking_Status, we'll use C() in the formula for correct handling.

    # Make sure column names are valid for formula (no spaces, special characters)
    # Let's assume our current column names are fine based on previous steps.

    # Construct the regression formula
    # Using C(Gender) and C(Smoking_Status) to treat them as categorical
    # Including LBXGLU and LBXTR as potential predictors of LBXGH
    formula = f'{dependent_variable} ~ Age + C(Gender) + BMXBMI + C(Smoking_Status) + Avg_Sleep_Hours + Vigorous_Rec_Min_Week + Avg_Total_Sugar_g + LBXGLU + LBXTR'

    # Check if all variables in the formula exist in the DataFrame
    required_cols = [dependent_variable] + [col.strip() for col in formula.split('~')[1].replace('C(', '').replace(')', '').replace('+', ' ').split()]
    missing_required_cols = [col for col in required_cols if col not in df_combined_all.columns]

    if missing_required_cols:
        print(f"❗ Error: The following variables required for the regression model were not found in df_combined_all: {missing_required_cols}. Please check preceding steps.")
    else:
        try:
            # Build the multiple linear regression model
            model = smf.ols(formula, data=df_combined_all).fit()

            # Print overall model fit statistics
            print("--- Multiple Linear Regression Model Fit ---")
            print(f"R-squared: {model.rsquared:.4f}")
            print(f"Adjusted R-squared: {model.rsquared_adj:.4f}")
            print(f"F-statistic: {model.fvalue:.4f}, P-value: {model.f_pvalue:.4f}")


            # Extract and display key metrics per predictor in a tidy table
            print("\n--- Key Regression Metrics per Predictor ---")
            results_df = pd.DataFrame({
                'Coefficient': model.params,
                'Standard Error': model.bse,
                'T-statistic': model.tvalues,
                'P-value': model.pvalues
            })

            # Format P-value for better readability
            results_df['P-value'] = results_df['P-value'].apply(lambda x: f"{x:.4f}")

            # Optionally, add significance stars (keeping this as it's standard practice and user asked about it)
            results_df['Significance'] = results_df['P-value'].apply(lambda x: '***' if float(x) < 0.001 else ('**' if float(x) < 0.01 else ('*' if float(x) < 0.05 else '')))

            display(results_df)

            print("\nDiscussion of Results:")
            print("The model fit statistics show the overall performance of the regression model.")
            print("- **R-squared**: Indicates the proportion of variance in HbA1c explained by all predictors combined.")
            print("- **Adjusted R-squared**: A modified R-squared that accounts for the number of predictors in the model, useful for comparing models.")
            print("- **F-statistic and P-value**: Test the overall significance of the model. A low p-value suggests that the model as a whole is statistically significant.")
            print("\nThe table below shows the key regression metrics for each predictor variable.")
            print("- **Coefficient**: Represents the estimated change in HbA1c for a one-unit increase in the predictor, holding other variables constant.")
            print("- **P-value**: Indicates the statistical significance of the predictor. A p-value < 0.05 typically suggests a statistically significant relationship.")
            print("- **Significance**: Stars indicate the level of statistical significance (*** < 0.001, ** < 0.01, * < 0.05).")
            print("Variables with statistically significant p-values are considered to have a significant association with HbA1c in this model.")
            print("The magnitude and sign of the coefficients indicate the strength and direction of these significant relationships.")


        except Exception as e:
            print(f"❌ Error building or running the regression model: {e}")


else:
    print("❗ 'df_combined_all' DataFrame not found. Please ensure preceding cells are run correctly.")
```
